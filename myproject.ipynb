```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
city_day = pd.read_csv("city_day.csv")

```

```python
city_day.iloc[0]                      #2015 Ahemdabad details
```

```python
city_day.head()
city_day.tail()
```

```python
city_day.describe
```

```python
lucknow_data = city_day.groupby('City').get_group('Lucknow')
```

```python
lucknow_data.tail(10)
```

```python
lucknow_data.plot(x='Date', y='PM2.5', kind="scatter", title="Date VS PM2.5 In LUCKNOW")
```

# After Seeing Above Plot We can Say that Lucknow's PM2.5 level mostly lies between 0 to 300. 

```python
lucknow_data.plot(x='Date', y='NO', kind="scatter", title="Date VS NO In LUCKNOW")
```

# After Seeing Above Plot We can say that Lucknow's NO level Increase after 2015.

```python
lucknow_data.plot(x='Date', y='NO2', kind="scatter", title="Date VS NO2 In LUCKNOW")
```

```python
lucknow_data.plot(x='Date', y='CO', kind="scatter", title="Date VS CO In LUCKNOW")
```

# After seeing above plot we can say that after 2015 CO decrease in Lucknow by maximum 5.

```python
lucknow_data.plot(x='Date', y='SO2', kind="scatter", title="Date VS SO2 In LUCKNOW")
```

# after seeing above plot we can say that SO2 hasn't change in lucknow while it touched  maximum 175 in 2015.

```python
lucknow_data.plot(x='Date', y='AQI', kind="scatter", title="Date VS AQI In LUCKNOW")
```
